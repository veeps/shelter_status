# Pulling tweets for Hurricanes

Here we'll be pulling tweets around 3 major hurricanes:

|Hurricane|Hashtag|Dates|
|:---|:---|:---|
|Hurricane Dorian|#hurricanedorian|08-24-2019 to 09-17-2019|
|Hurricane Florence|#hurricaneflorence|08-31-2018 to 09-25-2019|
|Hurricane Harvey|#hurricaneharvey|08-17-2017 to 09-09-2017|


Tweets are pulled from the start date of the incidence to one week after the hurricane ended.

```python
from twitterscraper import query_tweets
import datetime as dt
import pandas as pd
import pickle
pd.set_option('display.max_columns', 500)
```

## Pull tweets for Hurricane Dorian

```python
lang= 'english'
```

```python
limit = 30000
```

```python
tweets_dorian = query_tweets('hurricanedorian', begindate=dt.date(2019,8,24), enddate=dt.date(2019,9,17),
                     limit=limit, lang=lang)
```

```python
# convert to dataframe
tweets_dorian_df=pd.DataFrame(t.__dict__ for t in tweets_dorian)
```

```python
# pickle for later use
pickle.dump(tweets_dorian_df, open("../Data/dorian_tweets.pkl", "wb"))
```

### Filter for tweets based on "need"

```python
needs_dorian = tweets_dorian_df[tweets_dorian_df["text"].str.contains("need")]
```

```python
# Write to csv 
needs_dorian.to_csv("../Data/dorian.csv")
```
